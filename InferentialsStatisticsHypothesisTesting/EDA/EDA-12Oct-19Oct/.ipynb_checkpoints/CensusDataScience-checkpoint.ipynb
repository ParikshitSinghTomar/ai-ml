### number of literate people aged 7 or above / total population aged 7 or above

```python
# first of all find literate people aged 7 or above
# to do this Tatal people - people age between 0-6
df2['Literate_Persons'].values[0]
```

```python
#Total people aged 0-6
x=df[df['AgeGroup'].apply(lambda x: x=='0-6')][(df['AgeGroup']=='0-6') & (df['AreaType']=='Total') & (df['Area Name']=='INDIA')]['Literate_Persons'].values[0]
```

```python
y=df[df['AgeGroup'].apply(lambda x: x=='All ages')][(df['AgeGroup']=='All ages') & (df['AreaType']=='Total') & (df['Area Name']=='INDIA')]['Literate_Persons'].values[0]
```

```python
total_remaining=y-x
total_remaining
```

```python
# Total population aged 7 or above
x=df[df['AgeGroup'].apply(lambda x: x=='0-6')][(df['AgeGroup']=='0-6') & (df['AreaType']=='Total') & (df['Area Name']=='INDIA')]['TotalPersons'].values[0]
```

```python
x
```

```python
y=df[df['AgeGroup'].apply(lambda x: x=='All ages')][(df['AgeGroup']=='All ages') & (df['AreaType']=='Total') & (df['Area Name']=='INDIA')]['TotalPersons'].values[0]
```

```python
y
```

```python
remainingPersons=y-x
remainingPersons
```

```python
total_remaining
```

```python
(total_remaining/remainingPersons)*100
```
