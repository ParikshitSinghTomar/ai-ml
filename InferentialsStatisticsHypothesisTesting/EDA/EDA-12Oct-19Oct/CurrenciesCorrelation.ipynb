```python
import numpy as np
alphabets = ['Euro', 'Japanese Yen', 'U.K. Pound Sterling','U.S. Dollar','Australian Dollar','Indian Rupee']
  
figure = plt.figure(figsize=[7,5]) 
axes = figure.add_subplot(111) 
caxes = axes.matshow(fd2) 
figure.colorbar(caxes) 
  
axes.set_xticklabels(['']+alphabets) 
axes.set_yticklabels(['']+alphabets) 
  
plt.show() 
```

```python
f = plt.figure(figsize=(19, 15))
plt.matshow(fd2.corr(), fignum=f.number)
plt.xticks(range(fd2.shape[1]), fd2.columns, fontsize=14, rotation=45)
plt.yticks(range(fd2.shape[1]), fd2.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```
