```python
import pandas as pd
```

```python
df=pd.read_csv('odi-batting.csv')
df.head()
```

```python
df['Century']=df['Runs'].apply(lambda x: x>99)
```

```python
df.head()
```

```python
df['Century'].value_counts()
```

```python
df.groupby(['Player'])['Century'].agg(sum).reset_index().sort_values(by='Century',ascending=True)
```

```python
df['StrikRate']=(df['Runs']/df['Balls'])*100
```

```python
df.head()
```

```python
df.shape
```

```python
df['CenturyStrikRate']=df['Runs']>=100
```

```python
df.head()
```

```python
df=df[df['CenturyStrikRate']==True]
```

```python
df.head()
```

```python
df.shape
```

```python
Batsman_strike_rate=df[['Player','StrikRate']]
Batsman_strike_rate.sort_values(by="StrikRate",ascending=False).reset_index(drop=True)
```

```python
df.groupby(['Country'])['Century'].agg(sum).reset_index().sort_values(by='Century',ascending=False)
```

```python
df['Year']=df['MatchDate'].apply(lambda x:x[-4:])
```

```python
df.head()
```

```python
df=df[df['Country']=='India']
```

```python
df.head()
```

```python
df.groupby(['Year'])['Century'].agg(sum).reset_index().sort_values(by='Century',ascending=False)
```
