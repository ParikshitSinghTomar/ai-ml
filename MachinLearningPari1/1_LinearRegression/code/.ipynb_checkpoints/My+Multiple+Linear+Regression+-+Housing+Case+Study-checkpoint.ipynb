## Step 2: Preparing the data for Modeling

- Encoding:
    - Converting binary categorical variables to one and zeros    
    - Converting other categorical variables to dummy variables
- Spliting into train and test
- Rescaling of variables

```python
housing.mainroad.value_counts()
```

```python
#yes/no variables
varlist=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
#housing[varlist].head()
housing[varlist]=housing[varlist].apply(lambda x: x.map({'yes':1,'no':0}))
housing[varlist].head()
```

### Dummy Variables

```python
# Creating dummy variables for furnishing status
status=pd.get_dummies(housing['furnishingstatus'])
status.head()
```

Now, you don't need three columns. You can drop the `furnished` column, as the type of furnishing can be identified with just the last two columns where â€” 
- `00` will correspond to `furnished`
- `01` will correspond to `unfurnished`
- `10` will correspond to `semi-furnished`

```python
# Creating dummy variables for furnishing status
# dropping a redundent dummy variables
status=pd.get_dummies(housing['furnishingstatus'],drop_first=True)
status.head()
```

```python
# concat the dummy df with the original data
housing=pd.concat([housing,status],axis=1)
housing.head()
```

```python
housing=housing.drop('furnishingstatus',axis=1)
housing.head()
```

## Spliting data into train-test sets

```python
df_train,df_test=train_test_split(housing, train_size=0.7, random_state=100)
print(df_train.shape)
print(df_test.shape)
```

### Rescaling the Features 

As you saw in the demonstration for Simple Linear Regression, scaling doesn't impact your model. Here we can see that except for `area`, all the columns have small integer values. So it is extremely important to rescale the variables so that they have a comparable scale. If we don't have comparable scales, then some of the coefficients as obtained by fitting the regression model might be very large or very small as compared to the other coefficients. This might become very annoying at the time of model evaluation. So it is advised to use standardization or normalization so that the units of the coefficients obtained are all on the same scale. As you know, there are two common ways of rescaling:

1. Min-Max scaling 
2. Standardisation (mean-0, sigma-1) 

This time, we will use MinMax scaling.

### From discussion of video. Benefit 1. Make more interpretable and increase performance cost function in background. 

Which one we should use
Use min max scaling. because its takes care of outliers. If there were a data points outliers those datapoints mapped to 1 and other mapped between 0 and -1. While in standarisation data points can still have few higher values.

The advantage of Standardisation over the other is that it doesn't compress the data between a particular range as in Min-Max scaling. This is useful, especially if there are extreme data point (outlier). 

### Note : There is no effect of Min-Max scaling on binary variable (consist of 0 and 1).

```python
# x
# normalisation (x-xmin)/(xmax-xmin)
# standarisation (x-mu)/sigma
```

scaler.fit():  learns xmin and xmax
scaler.transform() : (x-xmin)/(xmax-xmin)
scaler.fit_transform() : do both the thing in a single shot

```python
#1. Instantiate an object
scaler=MinMaxScaler()

#create a list of numerical vars
num_vars=['area','bedrooms','bathrooms','stories','parking','price']

#2. fit on data
#scaler.fit
df_train[num_vars]=scaler.fit_transform(df_train[num_vars])
df_train.head()
```

```python
#you can check that all values lies between 0 and 1
df_train[num_vars].describe()
```

## Step 3: Training the Model

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_train.corr(),annot=True,cmap='YlGnBu')
plt.show()
```

```python
# X_train, y_train
y_train=df_train.pop('price')
X_train=df_train
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# add constant
X_train_sm=sm.add_constant(X_train['area'])

#create first model
lr=sm.OLS(y_train,X_train_sm)

#fit
lr_model=lr.fit()


#params
lr_model.params
```

```python
lr_model.summary()
```

### So what we can look 

1. coef(area):0.46 which looks fine
2. the most importent thing is significance,P-value. P-Value is 0 for both intercept and area, now we know that this resutlt is statistically significant. 
3. Now look at r-squared 0.283. It means we have that about 28% variance in price is explained by variable area. Which is not very bad but we need to improve the model. 

So we add another variable. 

There are two ways to create model.
1. Add variables one by one and find better variable 
2. remove variables which dont have good impact on price.

```python
# add another variable bathrooms(0.52 corr in heatmap)

#X_train_sm=X_train[['area','bathrooms']]
# add constant
X_train_sm=sm.add_constant(X_train[['area','bathrooms']])

#create first model
lr=sm.OLS(y_train,X_train_sm)

#fit
lr_model=lr.fit()


#params
lr_model.params
```

```python
lr_model.summary()
```

R-square value goes to 0.47 it mean 47% variance can be explained by area+bathrooms

```python
# add another variable badrooms(0.35 corr in heatmap)

# add constant
X_train_sm=sm.add_constant(X_train[['area','bathrooms','bedrooms']])

#create first model
lr=sm.OLS(y_train,X_train_sm)

#fit
lr_model=lr.fit()


#params
lr_model.summary()
```

### Note: if R-squared increase from 2% to 3% then its good to add variable
